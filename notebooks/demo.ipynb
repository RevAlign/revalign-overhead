# revalign-overhead: find pools from space, in one notebook

This runs the free, local pool detector over a real neighborhood in Paradise Valley, Arizona and shows you:

- an annotated proof image (green = confirmed pool, amber = tentative), and
- a georeferenced CSV (lat, lon, confidence, and a clickable Google Maps link per pool).

No API key, no cost. The imagery is free. Runtime is about 1 to 2 minutes (the model weights download once). Colab CPU is fine; just run the cells top to bottom.

## 2. Scan a neighborhood for pools
The free `yolo` backend runs entirely on this machine ($0 per scan).

```python
from revalign_overhead.detect import detect, OBJECTS

rows = detect(
    lat=33.5400, lon=-111.9500, size_m=510, z=19,
    spec=OBJECTS["pool"], backend="yolo", outdir="out_demo",
)
print(f"{len(rows)} detections written to out_demo/")
```

## 3. Look at the proof image
Green circles are confirmed pools (they passed the blue-water color gate); amber circles are tentative (kept, but flagged). Always eyeball this before trusting a list.

```python
from IPython.display import Image
Image("out_demo/detected.png")
```

## 4. The georeferenced results
One row per pool. Click any `gmaps` link to verify it on the map.

```python
import pandas as pd
df = pd.read_csv("out_demo/detections.csv")
print(df["status"].value_counts().to_dict())
df.head(20)
```
